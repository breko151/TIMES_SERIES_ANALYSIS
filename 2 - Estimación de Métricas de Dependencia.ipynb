Implemente las función de crossvariance descrita debajo

```python
def crossvariance(
    time_series_1: pd.Series,
    time_series_2: pd.Series,
    lags: Union[List[int], int]
) -> Union[List[float], float]:
    """
    Descripción:
    Esta funcion es la generalización de la autocovarianza.
    
    Entradas:
    - time_series_1: Primera serie de tiempo a utilizar para el cálculo.
    - time_series_2: Segunda serie de tiempo a utilizar para el cálculo.
    - lags: Puede ser uno o varios lags para los cuales debe calcularse la covarianza.
            
    Salidas:
    - Si lags es un número debe regresar un valor de covarianza, si es una lista, 
        debe ser una lista de covarianzas.
    """
    
    # CÓDIGO AQUI
    # Convertimos a arrays de np
    array_1 = np.array(time_series_1)
    array_2 = np.array(time_series_2)
    # Validamos si lags es un int o una lista
    if type(lags) == int:
        # Generamos nuestros arreglos con lags
        array_2_lags = array_2[lags:]
        array_1_lags = array_1[:len(array_1) - lags]
        # Obtenemos medias
        mean_array_1 = array_1_lags.mean()
        mean_array_2 = array_2_lags.mean()
        # Les restamos las medias
        array_1_lags_mean = array_1_lags - mean_array_1
        array_2_lags_mean = array_2_lags - mean_array_2
        # Hacmos operaciones entre vectores
        crossvariance_result = (1 / len(array_2_lags_mean)) * sum(array_1_lags_mean * array_2_lags_mean)
    else:
        # Inicializamos lista de resultados
        crossvariance_result = list()
        for lag in lags:
            # Inicializamos resultados temporales
            i_result = 0
            # Generamos arreglos con lags
            array_2_lags = array_2[lag:]
            array_1_lags = array_1[:len(array_1) - lag]
            # Obtenemos medias
            mean_array_1 = array_1_lags.mean()
            mean_array_2 = array_2_lags.mean()
            # Les restamos las medias
            array_1_lags_mean = array_1_lags - mean_array_1
            array_2_lags_mean = array_2_lags - mean_array_2
            # Hacemos operaciones entre vectores
            i_result = (1 / len(array_1_lags_mean)) * sum(array_1_lags_mean * array_2_lags_mean)
            crossvariance_result.append(i_result)
    return crossvariance_result
```
